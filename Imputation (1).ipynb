```
import pandas as pd
```

```
url = 'https://raw.githubusercontent.com/wblakecannon/DataCamp/master/10-cleaning-data-in-python/_datasets/airquality.csv'
df = pd.read_csv(url)
# Dataset is now stored in a Pandas Dataframe
```
