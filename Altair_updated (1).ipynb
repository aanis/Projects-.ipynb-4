```
# libraries

import pandas as pd
import altair as alt
```

```
url = 'https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/d546eaee765268bf2f487608c537c05e22e4b221/iris.csv'
df = pd.read_csv(url)
# Dataset is now stored in a Pandas Dataframe
```

```
df.head()
```

```
# Bar Chat

alt.data_transformers.enable('default', max_rows=None)

# plot the dataset, referencing dataframe column names
import altair as alt
chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('sepal_length'),
    y='count()',
      tooltip = 'sepal_length'
  
)

chart
```

```
#by color

# Bar Chat - Good for nominal and small ordinal categorical data.

alt.data_transformers.enable('default', max_rows=None)

# plot the dataset, referencing dataframe column names
import altair as alt
chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('sepal_length'),
    y='count()',
    color='sepal_length',
    tooltip = 'sepal_length'
)

chart
```

```
#Scatterplot with Filter


input_dropdown = alt.binding_select(options=['setosa','versicolor','virginica'])
selection = alt.selection_single(fields=['species'], bind=input_dropdown, name='Filter')


alt.data_transformers.enable('default', max_rows=None)

# plot the dataset, referencing dataframe column names
import altair as alt
chart = alt.Chart(df).mark_point().encode(
    x=alt.X('sepal_width'),
    y=alt.Y('sepal_length'),
    color='species',
    tooltip = 'species',
    
).add_selection(
    selection
).transform_filter(
    selection
)
 
chart
```

```
# Hisotgram - With Scales

scales = alt.selection_interval(bind='scales')


alt.data_transformers.enable('default', max_rows=None)

# plot the dataset, referencing dataframe column names
import altair as alt
chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('sepal_length',bin =True),
    y='count()',
    color='sepal_length',
    tooltip = 'sepal_length'
).add_selection(
    scales
)

chart
```

```
#Heatmap


scales = alt.selection_interval(bind='scales')


alt.data_transformers.enable('default', max_rows=None)

# plot the dataset, referencing dataframe column names
import altair as alt
chart = alt.Chart(df).mark_rect().encode(
    x=alt.X('sepal_length'),
    y=alt.Y('sepal_width'),
    color='sepal_length',
    tooltip = 'sepal_length',
    
).add_selection(
    scales
)

chart
```

```
#Line Chart




alt.data_transformers.enable('default', max_rows=None)

# plot the dataset, referencing dataframe column names
import altair as alt
chart = alt.Chart(df).mark_line(point=True).encode(
    x=alt.X('sepal_length'),
    y=alt.Y('sepal_width'),
    color='sepal_length',
    tooltip = 'sepal_length'
    

)

chart
```

```
#Strip Plot




alt.data_transformers.enable('default', max_rows=None)

# plot the dataset, referencing dataframe column names
import altair as alt
chart = alt.Chart(df).mark_tick(point=True).encode(
    x=alt.X('sepal_length'),
    y=alt.Y('sepal_width'),
    color='sepal_length',
    tooltip = 'sepal_length'
    

)

chart
```

```
#Area chart




alt.data_transformers.enable('default', max_rows=None)

# plot the dataset, referencing dataframe column names
import altair as alt
chart = alt.Chart(df).mark_area().encode(
    x=alt.X('sepal_length'),
    y=alt.Y('sepal_width'),
    color='species'
    

)

chart


```

```
#Boxplot




alt.data_transformers.enable('default', max_rows=None)

# plot the dataset, referencing dataframe column names
import altair as alt
chart = alt.Chart(df).mark_rule().encode(
    x=alt.X('sepal_length'),
    y=alt.Y('sepal_width'),
    color='species'
    

)

chart
```

```
#Heatmap binned


scales = alt.selection_interval(bind='scales')


alt.data_transformers.enable('default', max_rows=None)

# plot the dataset, referencing dataframe column names
import altair as alt
chart = alt.Chart(df).mark_rect().encode(
    x=alt.X('sepal_length', bin=alt.Bin(maxbins=60)),
    y=alt.Y('sepal_width', bin=alt.Bin(maxbins=60)),
    color='sepal_length',
    tooltip = 'sepal_length',
    
).add_selection(
    scales
)

chart
```
